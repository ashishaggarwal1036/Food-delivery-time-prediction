```python
import haversine as hs
Distance=[]
i=0
while i<data.shape[0]:
    a=hs.haversine((data["Restaurant_latitude"][i],data["Restaurant_longitude"][i]),(data["Delivery_location_latitude"][i],data["Delivery_location_longitude"][i]))
    Distance.append(a)
    i=i+1
data["Distance"]=np.round(Distance,2)
```

# Base Model

```python
X=new_data.drop("Time_taken(min)",axis=1)
y=new_data["Time_taken(min)"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
import statsmodels.api as sma
xtrain=sma.add_constant(xtrain)
xtest=sma.add_constant(xtest)
model=sma.OLS(ytrain,xtrain).fit()
model.summary()
```

## RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf_model=rf.fit(xtrain,ytrain)
```

```python
rf_pred=rf_model.predict(xtest)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
rf_r2=r2_score(ytest,rf_pred)
print(rf_r2)
```

```python
rf_mse=mean_squared_error(ytest,rf_pred)
print(rf_mse)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt_model=dt.fit(xtrain,ytrain)
```

```python
dt_pred=dt_model.predict(xtest)
```

```python
dt_r2=r2_score(ytest,dt_pred)
print(dt_r2)
dt_mse=mean_squared_error(ytest,dt_pred)
print(dt_mse)
```
